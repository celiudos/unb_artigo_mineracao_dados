# Carregar dados

```python
import pandas as pd

# # Mostrar mais colunas
pd.set_option("display.max_columns", 100)
pd.set_option('display.max_colwidth', 100)

def load_file(ano = 2019):
  path = f"./output/silver/rais_{ano}_prof_TI_RAIS_VINC_PUB.parquet.gzip"
  df = pd.read_parquet(path)

  qnt_total = len(df)

  print(f"""
  Quantidade docs: {qnt_total}
  """)

  return df

df_2019 = load_file(ano = 2019)
df_2019.head(2)
```

```python
df_2020 = load_file(ano = 2020)
df_2020.head(2)
```

```python
df_2021 = load_file(ano = 2021)
df_2021.head(2)
```

```python
df_2018 = load_file(ano = 2018)
df_2018.head(2)
```

# Somente colunas que importam

```python
RENAME_COLUNAS_QUE_IMPORTAM = columns = {
    'Município': 'id_uf_municipio',
    'Vl Remun Média Nom': 'remuneracao_media',
    'Motivo Desligamento': 'desligamento_id',
    'Idade': 'idade',
    'CBO Ocupação 2002': 'ocupacao_id',
    'Sexo Trabalhador': 'sexo_id'
}
```

```python
def somente_cols_importam(df, ano = 2019):
  df_r = df[['ano', 'Município', 'Vl Remun Média Nom', 'Motivo Desligamento', 'Idade',
             'CBO Ocupação 2002', 'Sexo Trabalhador']]

  df_r = df_r.rename(columns=RENAME_COLUNAS_QUE_IMPORTAM)

  df_r.head(2)
  return df_r

df_c_2019 = somente_cols_importam(df_2019, ano = 2019)
df_c_2019.head(2)
```

```python
df_c_2020 = somente_cols_importam(df_2020, ano = 2020)
df_c_2020.head(2)
```

# Juntando com UF

```python
def load_file_mun(df, ano = 2019):
  path =f"./output/silver/uf_municipio_{ano}.parquet.gzip"
  df_mun = pd.read_parquet(path)

  df_mun["id_uf_municipio"] = df_mun["id_uf_municipio"].astype(int)

  df_m = pd.merge(df, df_mun, on='id_uf_municipio', how='left', suffixes=('', '_mun'))
  df_m.drop(columns=['ano_mun'], inplace=True)

  qnt_total = len(df)

  print(f"""
  Quantidade docs: {qnt_total}
  """)

  return df_m

df_d_2019 = load_file_mun(df_c_2019, ano = 2019)
df_d_2019.head(2)
```

```python
df_d_2020 = load_file_mun(df_c_2020, ano = 2020)
df_d_2020.head(2)
```

```python
df_d_2018 = load_file_mun(df_c_2018, ano = 2019)
df_d_2018.head(2)
```

```python
df_d_2021 = load_file_mun(df_c_2021, ano = 2020)
df_d_2021.head(2)
```

# Juntando tudo

```python
df = pd.concat([df_d_2019, df_d_2020, df_d_2018, df_d_2021])
len(df)
```

```python
df["sigla_uf"].value_counts()
```

# Salvando arquivo

```python
df.to_parquet("./output/silver/rais_TODOS_ANOS_RAIS_VINC_PUB.parquet.gzip", index=False, compression="gzip")
```

## Arquivo com todas as colunas

```python
def somente_cols_importam_raw(df, ano=2019):

  df_r = df.rename(columns=RENAME_COLUNAS_QUE_IMPORTAM)

  df_r.head(2)
  return df_r
```

```python
df_e_2018 = somente_cols_importam_raw(df_2018, ano=2018)
df_e_2019 = somente_cols_importam_raw(df_2019, ano=2019)
df_e_2020 = somente_cols_importam_raw(df_2020, ano=2020)
df_e_2021 = somente_cols_importam_raw(df_2021, ano=2021)

df_f_2018 = load_file_mun(df_e_2018, ano=2019)
df_f_2019 = load_file_mun(df_e_2019, ano=2019)
df_f_2020 = load_file_mun(df_e_2020, ano=2020)
df_f_2021 = load_file_mun(df_e_2021, ano=2020)

df_g = pd.concat([df_f_2019, df_f_2020, df_f_2018, df_f_2021])

df_g.head(2)
```

```python
df_g["Mês Desligamento"] = df_g["Mês Desligamento"].replace("{ñ", -1).astype(int)
```

```python
# Salvando com todas as colunas
df_g.to_parquet("./output/silver/rais_TODOS_ANOS_TODAS_COLS_RAIS_VINC_PUB.parquet.gzip",
              index=False, compression="gzip")
```
