# Carregar dados

```python
ANO = 2017
```

```python
import pandas as pd

# # Mostrar mais colunas
pd.set_option("display.max_columns", 100)
pd.set_option('display.max_colwidth', 100)

# pandas read txt and decode utf-8 and read until 1000 rows

IDS_PROFISSIONAIS_TI = ["212205", "212210", "212215", "212305", "212310", "212315", "212320", "212405", "212410", "212415", "212420", "317105", "317110", "317115", "317120", "317205", "317210"]
OUTROS_PROFISSIONAIS = [
    "241005", # Advogado
    "252105", # Administrador
    "252210", # Contador
    "411010", # Assistente Administrativo
    "411005", # Auxiliar de Escritorio, em Geral
]

def load_csv(path):
    dtype = {
      'CBO Ocupação 2002': 'string',
      'Motivo Desligamento': int
    }
    
    df = pd.read_csv(
        path, 
        encoding = "ISO-8859-1", 
        sep = ";", 
        decimal = ",", 
        dtype = dtype
        )

    qnt_total = len(df)

    df = df[df['CBO Ocupação 2002'].isin(IDS_PROFISSIONAIS_TI + OUTROS_PROFISSIONAIS)] 

    print(f"""
    Path: {path}
    Quantidade docs: {qnt_total}
    Quantidade docs filtrados: {len(df)}
    """)

    return df

# df_estados = load_csv(f"./output/bronze/rais_{ANO}/DF{ANO}.txt")
```

```python
ESTADOS = ["AC", "AL", "AM", "AP", "BA", "CE", "DF", "ES", "GO", "MA", "MG", "MS", "MT", "PA", "PB", "PE", "PI", "PR", "RJ", "RN", "RO", "RR", "RS", "SC", "SE", "SP", "TO"]

df_estados = [load_csv(f"./output/bronze/rais_{ANO}/{e}{ANO}.txt") for e in ESTADOS]
len(df_estados)
```

# Juntando

```python
df_tudo = pd.concat(df_estados)
df_tudo
```

```python
df_tudo['CBO Ocupação 2002'].value_counts()
```

```python
df_tudo["ano"] = ANO
df_tudo.head(2)
```

```python
df_tudo["Bairros RJ"].value_counts()
```

# Salvando arquivo

```python
df_tudo["Mês Desligamento"] = df_tudo["Mês Desligamento"].replace("{ñ", -1)
df_tudo["Mês Desligamento"] = df_tudo["Mês Desligamento"].astype(int)

df_tudo["Bairros RJ"] = df_tudo["Bairros RJ"].replace('           {ñ class}', -1).replace("{ñ c", -1)
df_tudo["Bairros RJ"] = df_tudo["Bairros RJ"].astype(int)

df_tudo["Distritos SP"] = df_tudo["Distritos SP"].replace('           {ñ class}', -1)
df_tudo["Distritos SP"] = df_tudo["Distritos SP"].astype(int)
```

```python
df_tudo.to_parquet(f"./output/silver/rais_{ANO}_prof_TI_RAIS_VINC_PUB.parquet.gzip", index=False, compression="gzip")
```
